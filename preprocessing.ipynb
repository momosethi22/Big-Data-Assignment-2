```python
import pandas as pd
import string
import re

# Specify the path to your CSV file
csv_file_path = '/home/momo/Downloads/enwiki-20170820.csv'

# Define chunk size
chunk_size = 10000  # You can adjust this based on your system's memory capacity

# Define preprocessing function
def preprocess_text(text):
    # Check if the value is NaN
    if pd.isna(text):
        return ''  # Return empty string if NaN
    # Convert to lowercase
    text = str(text).lower()
    # Remove non-alphanumeric characters and symbols
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Remove newline characters
    text = text.replace('\n', ' ')
    # Remove extra whitespaces
    text = ' '.join(text.split())
    return text

# Read the CSV file in chunks
chunk_iter = pd.read_csv(csv_file_path, chunksize=chunk_size)

# Process each chunk and write to file
with open('data.txt', 'w') as f:
    for i, chunk in enumerate(chunk_iter):
        print(f"Processing chunk {i+1}")
        # Drop specified columns
        chunk.drop(['TITLE', 'SECTION_TITLE'], axis=1, inplace=True)
        # Apply preprocessing to 'SECTION_TEXT' column
        chunk['SECTION_TEXT'] = chunk['SECTION_TEXT'].apply(preprocess_text)
        # Write to file
        for index, row in chunk.iterrows():
            f.write(f"{row['ARTICLE_ID']}\t{row['SECTION_TEXT']}\n")

print("Preprocessing completed.")
```
